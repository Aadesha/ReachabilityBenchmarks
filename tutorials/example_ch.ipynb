```julia
using LazySets
```

```julia
A = sprandn(1000, 1000, 0.01)
δ = 0.2
expA = SparseMatrixExp(A * δ)
X0 = Ball2(ones(1000), 0.01)
B = randn(1000, 2)
U = BallInf(zeros(2), 0.7);
```

```julia
@time M = expm(full(A));
```

```julia
@time Y = CH(expA * X0 + δ * B * U, X0);
```

```julia
typeof(Y)
```

```julia
proj_mat = [[1. zeros(1, 999)]; [zeros(1, 499) 1. zeros(1, 500)]]
```

```julia
@time res1 = Approximations.overapproximate(proj_mat * Y);
```

```julia
using Plots
```

```julia
p1 = plot(res1, aspect_ratio=:equal, label="Axis-aligned box")
```

```julia
@time res2 = Approximations.overapproximate(proj_mat * Y, 1e-1);
```

```julia
p2 = plot(res2, aspect_ratio=:equal, label="Refined: 1e-1")
```

```julia
@time res3 = Approximations.overapproximate(proj_mat * Y, 1e-3);
```

```julia
p3 = plot(res3, aspect_ratio=:equal, label="Refined: 1e-3")
```

```julia
pall = plot(p1, p2, p3, layout=(1, 3), legend=false)
```

```julia
savefig(pall, "example_ch.png")
```
